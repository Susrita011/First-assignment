```python
#11. Identify and remove outliers in the km_driven column using the ICR method ? 
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 -Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

filtered_df = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

print(filtered_df)
```

```python
# 13. what is the average depreciation in selling price based on the bike's age (current year - manufacturing
#year) ?

current_year = 2025
df['bikes_age'] = current_year - df['year']

avg_price_by_age = df.groupby('bikes_age')['selling_price'].mean().reset_index()

avg_price_by_age['depriciation'] = avg_price_by_age['selling_price'].diff()
print(avg_price_by_age)
```

```python
# 14. which bike names are priced significantly above the average price for their manufacturing year?

# Step 1: Group by year and calculate average price and standard deviation
year_stats = df.groupby('year')['selling_price'].agg(['mean', 'std']).reset_index()
year_stats.rename(columns={'mean': 'average_price', 'std': 'price_std'}, inplace=True)

# Step 2: Merge statistics back to the original DataFrame without duplicate column names
df = df.merge(year_stats, on='year', suffixes=('', '_y'))

# Step 3: Define the threshold (1 standard deviation above the average)
df['above_threshold'] = df['selling_price'] > (df['average_price'] + df['price_std'])

# Step 4: Filter bikes priced significantly above average
significantly_priced_bikes = df[df['above_threshold']][['name', 'year', 'selling_price']]

# Display results
print(significantly_priced_bikes)
```
