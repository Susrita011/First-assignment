4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate a dataset with categories and numerical values
categories = ['Category A', 'Category B', 'Category C', 'Category D']
data = {
    'Category A': np.random.normal(50, 10, 100),
    'Category B': np.random.normal(60, 15, 100),
    'Category C': np.random.normal(55, 20, 100),
    'Category D': np.random.normal(65, 5, 100)
}

# Prepare data for box plot
data_values = [data[category] for category in categories]

# Create a box plot to visualize the distribution across categories
plt.figure(figsize=(12, 8))
plt.boxplot(data_values, labels=categories, patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='blue'),
            medianprops=dict(color='red'))

# Add titles and labels
plt.title('Distribution of Numerical Variable Across Different Categories')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(42)
n_samples = 100
x1 = np.random.randn(n_samples)
x2 = x1 + np.random.randn(n_samples) * 0.5
x3 = x2 + np.random.randn(n_samples) * 0.2

data = pd.DataFrame({'Feature1': x1, 'Feature2': x2, 'Feature3': x3})

# Calculate the correlation matrix
corr_matrix = data.corr()

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

4. Using the given x and y data, generate a 3D surface plot to visualize the function(z = sin(sqrt(x**2 + y**2)))

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate z values
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z)])

# Customize the plot (optional)
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x**2 + y**2))',
                  scene=dict(xaxis_title='x',
                             yaxis_title='y',
                             zaxis_title='z'))

fig.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Ensure the plot displays in the notebook
output_notebook()

# Generate x and y values
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the results
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Ensure the plot displays in the notebook
output_notebook()

# Generate random data
np.random.seed(123)
N = 100
x = np.random.rand(N) * 10
y = np.random.rand(N) * 10
sizes = np.random.randint(10, 20, size=N)
colors = ["#"+''.join([np.random.choice(list('ABCDEF0123456789')) for i in range(6)]) for j in range(N)]

# Create a DataFrame for easy handling
data = {'x': x, 'y': y, 'sizes': sizes, 'colors': colors}
df = pd.DataFrame(data)

# Create a new plot
p = figure(title="Scatter Plot with Random Data", x_axis_label='x', y_axis_label='y')

# Add scatter plot with varying sizes and colors
p.scatter(x='x', y='y', size='sizes', color='colors', source=df)

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Ensure the plot displays in the notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Create histogram bins
hist, edges = np.histogram(data_hist, bins=30)

# Create a new plot
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

# Add a histogram renderer
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

# Show the plot
show(p)
```
