```python
# 2. which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing
#trends ?

brand_price_variation = df.groupby('Company')['Price ($)'].std().sort_values(ascending = False)

highest_variation_brand = brand_price_variation.idxmax()
highest_variation_price = brand_price_variation.max()
print(f" The highest variation price is {highest_variation_price} for the car :{highest_variation_brand}.")
```

```python
# Summary statistics for price variation across brands
brand_price_variation_summary = brand_price_variation.describe()
brand_price_variation_summary
```

```python
# 3. what is the distribution of car prices for each transmission type, and how do the interquartile ranges
# compare?

transmission_price_stats = df.groupby('Transmission')['Price ($)'].describe()
# Extract interquartile range (IQR) for each transmission type
transmission_price_stats['IQR'] = transmission_price_stats['75%'] - transmission_price_stats['25%']

transmission_price_stats
```

```python
# 5. what is the distribution of cars based on body styles ?

# Calculate the distribution of cars based on body styles
count = df['Body Style'].value_counts()

percentage = df['Body Style'].value_counts(normalize = True) * 100

body_style_distribution_summary = pd.DataFrame({'Count':count,'Percentage':percentage}).sort_values(by='Count', ascending=False)

body_style_distribution_summary
```

```python
# Calculate the distribution of cars based on body styles
body_style_distribution = df['Body Style'].value_counts()

# Add percentages for each body style
body_style_distribution_percentage = (body_style_distribution / body_style_distribution.sum()) * 100

# Combine the counts and percentages into a single DataFrame
body_style_distribution_summary = pd.DataFrame({
    'Count': body_style_distribution,
    'Percentage': body_style_distribution_percentage
}).sort_values(by='Count', ascending=False)

body_style_distribution_summary
```

```python
# 7. what is the distribution of car prices by region, and how does the number of cars sold vary by region?

region_price_distribution = df.groupby('Dealer_Region')['Price ($)'].describe()

car_sold_by_region = df['Dealer_Region'].value_counts()

print("Distribution of Car Prices by Region:")
print(region_price_distribution)

print("\nNumber of Cars sold by region:")
print(car_sold_by_region)
```

```python
# Create income ranges
income_bins = [0, 50000, 100000, 200000, 500000, 1000000, float('inf')]
income_labels = ['<50k', '50k-100k', '100k-200k', '200k-500k', '500k-1M', '>1M']
df['Income Range'] = pd.cut(df['Annual Income'], bins=income_bins, labels=income_labels)

# Calculate average selling price by gender
average_price_by_gender = df.groupby('Gender')['Price ($)'].mean()

# Calculate average selling price by gender and income range
average_price_by_gender_income = df.groupby(['Gender', 'Income Range'])['Price ($)'].mean()

average_price_by_gender, average_price_by_gender_income
```

```python
# 9.how do car prices vary based on the customer’s annual income bracket ?
# Define income brackets
income_bins = [0, 50000, 100000, 200000, 500000, 1000000, float('inf')]
income_labels = ['<50k', '50k-100k', '100k-200k', '200k-500k', '500k-1M', '>1M']
df['Income Bracket'] = pd.cut(df['Annual Income'], bins=income_bins, labels=income_labels)

# Group by income bracket and calculate price statistics
price_by_income_bracket = df.groupby('Income Bracket')['Price ($)'].describe()

# Display the results
print("Car Prices by Customer Annual Income Bracket:")
print(price_by_income_bracket)
```

```python
# 10. what are the top 5 car models with the highest number of sales, and how does their price distribution look?

# Identify the top 5 car models with the highest sales
top_models = df['Model'].value_counts().head(5)

# Filter data for the top 5 models
top_models_data = df[df['Model'].isin(top_models.index)]

# Analyze price distribution for the top 5 models
price_distribution = top_models_data.groupby('Model')['Price ($)'].describe()

# Display the results
print("Top 5 Car Models by Sales:")
print(top_models)

print("\nPrice Distribution for Top 5 Car Models:")
print(price_distribution)
```

```python
# 11. how does car price vary with engine size across different car colors, and which colors have the highest price
#variation?

# Group by Engine and Color, then calculate average price
price_by_engine_color = df.groupby(['Engine','Color'])['Price ($)'].mean().unstack()

# Calculate price variation (standard deviation) for each color
color_price_variation = df.groupby('Color')['Price ($)'].std().sort_values(ascending = False)

# Display results
print("Car Price by Engine Size and Color:")
print(price_by_engine_color)

print("\nPrice Variation by Color:")
print(color_price_variation)
```

```python
# 12. Is there any seasonal trend in car sales based on the date of sale ?

# Ensure 'Date of Sale' is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract month and season
df['Month'] = df['Date'].dt.month
df['Season'] = df['Month'].map({
    12: 'Winter', 1: 'Winter',2:'Winter',
    3:'Spring', 4:'Spring', 5: 'Spring',
    6:'Summer' , 7: 'Summer' , 8: 'Summer',
    9: 'Fall' , 10: 'Fall' ,11: 'Fall'
})

sales_by_month = df['Month'].value_counts().sort_index()
sales_by_season = df['Season'].value_counts()

# Display results
print("Car Sales by Month:")
print(sales_by_month)

print("\nCar Sales by Season:")
print(sales_by_season)
```

```python
# 14. what is the correlation between car price, engine size, and annual income of customers, and how do these
# features interact ?

# Ensure numerical representation of Engine
df['Engine Size'] = df['Engine'].str.extract(r'(\d+\.?\d*)').astype(float)

# Select relevant columns for analysis
features = df[['Price ($)', 'Engine Size', 'Annual Income']]

# Calculate correlations
correlation_matrix = features.corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Between Price, Engine Size, and Annual Income")
plt.show()

# Scatter plot to visualize interactions
sns.pairplot(features)
plt.show()
```
